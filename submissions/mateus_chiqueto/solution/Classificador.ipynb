```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import f1_score

# 1. Carregar os dados
df = pd.read_csv('merged_tickets.csv')

# Remover possíveis nulos
df = df.dropna(subset=['texto', 'categoria'])

X = df['texto']
y = df['categoria']

# 2. Dividir em 70% treino e 30% teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Criar e treinar o pipeline (TF-IDF + Regressão Logística)
modelo = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('lr', LogisticRegression(max_iter=1000))
])
modelo.fit(X_train, y_train)

# 4. Avaliar com F-score
y_pred = modelo.predict(X_test)
f_score = f1_score(y_test, y_pred, average='weighted')

print(f"F-score: {f_score:.4f}")

# Exemplo de uso para prever um novo texto
novo_texto = ["My laptop is not turning on"]
categoria_predita = modelo.predict(novo_texto)
print(f"Categoria predita: {categoria_predita[0]}")
```
