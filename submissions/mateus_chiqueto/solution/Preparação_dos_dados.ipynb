```python
import pandas as pd

# Carregar os arquivos
df_customer = pd.read_csv('customer_support_tickets.csv')
df_processed = pd.read_csv('all_tickets_processed_improved_v3.csv')

# Manter apenas texto e categoria, padronizando os nomes das colunas
df1_clean = df_customer[['Ticket Description', 'Ticket Type']].rename(columns={
    'Ticket Description': 'texto',
    'Ticket Type': 'categoria'
})

df2_clean = df_processed[['Document', 'Topic_group']].rename(columns={
    'Document': 'texto',
    'Topic_group': 'categoria'
})

# Unir os dois dataframes
df_final = pd.concat([df1_clean, df2_clean], ignore_index=True)

# Remover linhas vazias (opcional, mas recomendado)
df_final.dropna(subset=['texto', 'categoria'], inplace=True)

# Salvar o novo arquivo
df_final.to_csv('merged_tickets.csv', index=False)
```
